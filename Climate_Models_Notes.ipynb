## Notes for Climate Models
OLR : Outgoing Longwave Radiation, (pretty stait foreward what this is) is observed to be $396\ \rm W m^{-2}$ calculated with;
$$OLR = σT^{4}_{e}$$
or rearranged, we can find the ;
$$T_e = (\frac{OLR}{σ})^{\frac{1}{4}}$$

```python
import numpy as np
from scipy.constants import sigma

OLR = 238.5
T_e = ( OLR /sigma)**(1/4)
olr = sigma * T_e**4

print(sigma)
print(T_e)
print(olr)
```

More precisely,
$$OLR = \tau \sigma T_e^4$$
Where $\tau$ is the transmistivity of the atmosphere. For earth, τ is less than one. This is because of the 'greenghouse gas effect' of the atmosohere. Where the OLR < ILR (ILR is Incoming Longwave Radiation). We can tune this specific model to be $\tau ≈ 0.61$.

More precisely;

```python
OLRobserved = 238.5  # in W/m2
sigma = 5.67E-8  # S-B constant
Tsobserved = 288.  # global average surface temperature
tau1 = OLRobserved / sigma / Tsobserved**4  # solve for tuned value of transmissivity
print(tau1)
```

Now we can take into account the Shortwave (solar) radiation
##Asorbed Shortwave Radiation (ASR)##
Terms to define
- Global Mean Insolation
  - The area-avaerage incoming solar radiation or **insolation** is $341.3 \ \rm W m^-2$ and is denoted by the letter **Q**.
- Planetary Albedo
  - Some of the incoming radiation is nor absorbed but simply reflected back into space. Name is $F_{reflected}$. We observe this to be (for Earth) $F_{reflected} = 101.9\ \rm Wm^-2$.
  - Note: **planetary albedo** is the fraction of **Q** that is reflected. We will call this **planetary albedo** or α.

  So far we have all of this;



```python
Q = 341.3 # in Wm^-2
F_reflected = 101.9 # in Wm^-2

alpha1 = F_reflected / Q

print(alpha1)
```

Therefore, about a third of the Shortwave Radiation is reflected back into space.

## Absorbed Shortwave Radiation

The absorbed Shortwave Radiation (ASR) is the the part that is **not** reflected. i.e. the part that is absorbed.

We can find this through;

$$ASR = Q - F_{reflected} = (1-α)Q$$


```python
ASR = Q - F_reflected

print(ASR)
```

## Equalibrium

The Earth System is said to be in equalibrium when
$$OLR = ASR$$

Entering the terms taht we have defined above, we find that; (Note that here we changed $T_e → T_s$ bu this is just a matter of terminology)

$$\tau \sigma T_s^4 = Q - F_{reflected}$$
$$\text{OR}$$
$$\tau \sigma T_s^4 = (1 - \alpha)Q$$

Solving for $T_s$ we obtian;

$$T_s^4 = \frac{(1-\alpha)Q}{\tau \sigma}$$

Taking the fourth root, we obtian our equalibruim temperature $T_{eq}$;

$$T_{eq} = \left( \frac{(1-\alpha)Q}{\tau \sigma} \right)^{\frac{1}{4}}$$

Run this through some code and we obtain;

```python
bert1 = tau1*sigma
merp1 = (1-alpha1)*Q

T_eq1 = (merp1/bert1)**(1/4)

print(T_eq1)
```

#4. Climate Change Scenario

What if, due to global warming, these things happen?
* The lingwave transmissivity decreases to $\tau = 0.57$
* The planetary albedo increases to $\alpha = 0.32$

Then what will happen to our $T_{eq}$?
<br>
We can re-run our numbers to find the new $T_{eq}$

```python
tau2 = 0.57
alpha2 = 0.32

bert2 = tau2*sigma
merp2 = (1-alpha2)*Q

T_eq2 = (bert2/merp2)**(1/4)
diff = T_eq1 - T_eq2

print(T_eq2)
print("The difference in equalibrium temperature is ")
print(diff)
```

## Chapter 7 Notes

1. Single layer atmosphere
  - Single layer with temp $T_a$
  - Completely transparent to shortwave solar radiation
  - The **surface** absorbs shortwave radiation $(1-\alpha)Q$, this is the absorbed shortwave radiaton equation from earlier. Where α is the $\frac{F_{reflected}}{Q}$ and $Q$ is the Global Mean Insolation or the area-average incoming solar radiation. Therfore, $(1 - \alpha)$ is the part that is **not reflected**.
  - The atmosphere is completely opaque to infrared radiation
2. Energy balance at the surface
$$energy \text{ } in = energy \text{ } out$$
$$(1 - \alpha)Q + \sigma T^4_a = \sigma T_s^4$$
T_a is the temp of the atmosphere, and T_s is th temp of the surface. Notice that for the surface side, there is no energy out for the term describing the shortwave radiation. This is because (as we said earlier) the short wave is absorbed by the surface. Intuition tells us that this will lead to the surface begin hotter than the atmosphere. So, in order to have a truly **radiative equalibrium**, he surface has to be hotter than the atmosphere.

3. Energy Balance for the atmosphere
$$energy \text{ } in = energy \text{ } out$$
$$\sigma T^4_s = 2 \sigma T^4_a$$

From these equations, we obtain;
$$T_s = 2^{\frac{1}{4}}T_a ≈ 1.2T_a$$
 - solve for the radiative equalibrium surface temp by plugging in the surface equation. Note that $T_e$ is the same as $T_a$ so we can substitute from $(1 - α)Q$ out for $\sigma T_e^4$
$$\frac{1}{2} T_s^4 = (1-α)Q$$
$$\frac{1}{2} T_s^4 = T_e^4$$
Now, solve for $T_s$
$$T_s = 2^{\frac{1}{4}}T_e$$
If we know from observations that $T_e = 255 K$ we can say that $T_s = 303$

However, this is an overprediction. The reason for this has to do with our initial assumptions. It is true that the atmosphere absorbs some of the longwave radiation incident upon it , however, it doesn't absorb **all** of it. If we set up a 2 layer atmosphere, where the absorbtivity (ϵ) is the same in each layer but not equal to 0, then we can a have a model closer to real life. We think of this new model formally as the Grey Gas Model and informally as the "leaky greenhouse".

Note: the 2 most importnant absorbers are
* $CO_2$ (well mixed)
* $H_2O$ (not well mixed, mainly stays in the lower troposphere due to strong temp dependence of the saturatin vapor pressure)
[This is ignored for rn]

## Kirchoff's Law
$$absorbtivity = emissivity$$
Therefore, if a layer of atmosphere at temp $T$ absorbs some fraction of longwave radiation incident upon it, it must emit $ϵσT^4$ both up and down.

[Look at the 2layerAtm_sketch.png in downloads for the next part]


- From the surface to Layer 1 of the atmo. the upward flux is just the emission from the surface;
$$U_0 = σT^4_s$$
- From layer 1 to layer 2, the emssion depends on the fraction of the beam that is **not** absorbed by layer 0 and the transmitted part of whatever is incident from below. The first part is dependent if ϵ and is written as follows; $(1-ϵ)σT^4_s$. And the second part is written as; $ϵσT^4_0$. These combined will give us;
$$U_1 = (1-ϵ)σT^4_s + ϵσT^4_0$$.

This same logic follows for the nect layers, if ther were any.

Tuning this model to observations in provided in the text

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.constants import sigma

e = np.arange(0,1.01,0.01)
T_s = 288
T_0 = 275
T_1 = 230

term1 = ((1 - e)**2)*sigma*(T_s)**4
term2 = e*(1 - e)*sigma*(T_0)**4
term3 = e*sigma*(T_1)**4

OLR = term1 + term2 + term3

#print(OLR)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot(e,OLR,'r-',lw=3) #Note the order of x and y (x IS time)
ax.grid()

plt.title('OLR vs Epsilon', loc='center',fontsize=20)
ax.set_xlabel("Epsilon",fontsize=16)
ax.set_ylabel("OLR",fontsize=16)

plt.yticks(np.arange(150, 400, 15))

#from this we estimate a value for epsilon of e = 0.58 which is close the actual value of 0.586
```

```python
import numpy as np
from scipy.constants import sigma

e = 0.586
T_s = 288
T_0 = 275
T_1 = 230

OLR_s = ((1 - e)**2)*sigma*(T_s**4)
OLR_0 = e*(1 - e)*sigma*(T_0**4)
OLR_1 = e*sigma*(T_1**4)

OLR_T = OLR_s + OLR_0 + OLR_1

print("The value for the OLR contributing from the surface is " + str(OLR_s))
print("The value for the OLR contributing from the first layer of the atmosphere is " + str(OLR_0))
print("The value for the OLR contributing from the second layer of the atmosphere is " + str(OLR_1))
print("Adding these values up, we get " + str(OLR_T))
```

The top layer of the atmosphere is the single biggest contributor to OLR (at 92 $W/m^2$).

## Adding absorbers to this model
When adding absorbing greenhouse gasses to this model, it will change things. Think about how ϵ will change now that there are different absorbing gasses present. It will increase the fraction of incident longwave radiation absorbed in each layer. So ϵ increases when greenhouse gasses are added.

If we had ϵ to start with, then after adding greenhouse gasses, we will have $ϵ_{new} = ϵ + Δϵ$

```python
!pwd
%cd /content/Data/

import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
import matplotlib.colors as colors
import matplotlib.cm as cm

#plt.cm.seismic
vmin, vmax = -110,110
my_cmap = plt.colormaps['seismic']# cm.get_cmap('seismic')
norm = colors.Normalize(vmin,vmax)
cmmapable = cm.ScalarMappable(norm,my_cmap)
cmmapable.set_array(range(0,1))

season_idx = [0,92,180,272]
season_lbl = ["season %i" % i for i in range(1,5)]
obl_idx = [10,23,80]

yi_c = np.arange(-89,91,2)
xi_c = np.arange(0,360,4)
fs = 'large'

fig = plt.figure(figsize=(10,15),dpi=150)
ax11 = fig.add_subplot(321)
ax11.set_xticks([])
ax12 = fig.add_subplot(322)
ax12.set_xticks([])
ax12.set_yticklabels([])
ax21 = fig.add_subplot(323)
ax21.set_xticks([])
ax22 = fig.add_subplot(324)
ax22.set_xticks([])
ax22.set_yticklabels([])
ax31 = fig.add_subplot(325)
ax32 = fig.add_subplot(326)
ax32.set_yticklabels([])

ax_list = [ax11, ax12, ax21, ax22, ax31, ax32]
obl_val= [10,23,80]

for ax_i in range(0,3):
  ax1, ax2 = ax_list[2*ax_i],ax_list[2*ax_i+1]
  lat_p,days_p,temp_p = np.genfromtxt('Obliqities_contourplot_data[%i].txt' % obl_idx[ax_i], delimiter = ',', comments = '#',unpack=True)
  zi_c = griddata((days_p,lat_p),temp_p,(xi_c[None,:],yi_c[:,None]),method='nearest')
  Tmax = 110; Tmin = -Tmax; delT = 5
  clevels = np.arange(Tmin, Tmax+delT, delT)
  #cax = ax.contourf(xi, yi, zi, levels=clevels, cmap=my_cmap, vmin=Tmin, vmax=Tmax)
  #cbar = plt.colorbar(cax)
  ax1.contourf(xi_c,yi_c,zi_c,vmin=np.min(temp_p),vmax=np.max(temp_p), cmap=my_cmap, levels = clevels)
  #xi = data[0] = latitidues, yi = data[1] = days of the year, zi = data[2] = temperatures

  if ax_i == 0:
    cax = fig.add_axes([0.125,0.89,0.38,0.015])
    cbar = plt.colorbar(cmmapable,cax=cax,orientation='horizontal',location='top')
    #cbar.ax.xaxis.set_ticks_position('top')
    cbar.set_label("Temperature $^\circ C$",fontsize=fs)

  for s in season_idx:
    s_slice = np.where(np.abs(days_p-s)<1e-6)[0]
    xvals = lat_p[s_slice] #latitudes for each season
    yvals = temp_p[s_slice] #temperatures for a given season with latitude
    ax2.plot(yvals,xvals,'-',lw=2,label=season_lbl[season_idx.index(s)])
    ax2.set_xlim(-75,105)
    ax2.set_ylim(-90,90)
  ax2.legend(loc='lower right',fontsize=10)
  ax1.set_ylabel("Latitude ($^\circ$)",fontsize=fs)
  ax1.text(345,80,"$\epsilon = %i^\circ$" % obl_val[ax_i],color='w',fontweight='bold',horizontalalignment='right')
  ax2.text(100,80,"$\epsilon = %i^\circ$" % obl_val[ax_i],color='k',fontweight='bold',horizontalalignment='right')
  if ax_i == 2:
    ax1.set_xlabel("Number of days",fontsize=fs)
    ax2.set_xlabel("Temperature ($^\circ C$)")

plt.subplots_adjust(wspace = 0.05, hspace = 0.01)
fig.savefig("Contour_plots_together.png",bbox_inches='tight',dpi=150)

#this is starting to make sense!!!
```
